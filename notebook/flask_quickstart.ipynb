6. **세션 관리** : Flask의 세션을 사용하여 사용자 정보를 유지하고 인증을 관리합니다.

```python

from flask import Flask, session, redirect, url_for
app = Flask(__name__)
app.secret_key = 'mysecretkey'

@app.route('/login')
def login():
    session['user'] = 'john'
    return 'Logged in successfully'

@app.route('/logout')
def logout():
    session.pop('user', None)
    return redirect(url_for('login'))

@app.route('/profile')
def profile():
    if 'user' in session:
        return f'Welcome, {session['user']}'
    else:
        return redirect(url_for('login'))

if __name__ == '__main__':
    app.run()
```

 
7. **데이터베이스 연동** : Flask와 SQLAlchemy를 사용하여 데이터베이스와 상호 작용합니다.

```python

from flask import Flask
from flask_sqlalchemy import SQLAlchemy

app = Flask(__name__)
app.config['SQLALCHEMY_DATABASE_URI'] = 'sqlite:///database.db'
db = SQLAlchemy(app)

class User(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    name = db.Column(db.String(50))

@app.route('/users')
def users():
    users = User.query.all()
    return ', '.join([user.name for user in users])

if __name__ == '__main__':
    app.run()
```

 
8. **폼 처리** : Flask-WTF를 사용하여 폼을 생성하고 제출한 데이터를 처리합니다.

```python

from flask import Flask, render_template, request
from flask_wtf import FlaskForm
from wtforms import StringField, SubmitField

app = Flask(__name__)
app.config['SECRET_KEY'] = 'mysecretkey'

class NameForm(FlaskForm):
    name = StringField('Name')
    submit = SubmitField('Submit')

@app.route('/', methods=['GET', 'POST'])
def form():
    form = NameForm()
    if form.validate_on_submit():
        name = form.name.data
        return f'Hello, {name}!'
    return render_template('form.html', form=form)

if __name__ == '__main__':
    app.run()
```

10. **인증과 권한 부여** : Flask-Login을 사용하여 사용자 인증과 권한 부여를 처리합니다.

```python
from flask import Flask
```

```python
from flask import Flask, render_template, request, redirect, url_for, session

app = Flask(__name__)
app.secret_key = 'secret_key'

    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        if username in users and users[username] == password:
            session['username'] = username
            return redirect(url_for('home'))
        else:
            return render_template('login.html', error='Invalid username or password.')
    return render_template('login.html')

# 로그아웃
@app.route('/logout')
def logout():
    session.pop('username', None)
    return redirect(url_for('login'))

# 인증된 사용자만 접근 가능한 홈 페이지
@app.route('/home')
def home():
    if 'username' in session:
        return render_template('home.html', username=session['username'])
    return redirect(url_for('login'))

if __name__ == '__main__':
    app.run(debug=True)
```

- Flask에서 `render_template` 함수가 템플릿 파일을 찾지 못하는 경우, 주로 템플릿 파일이 위치한 경로를 확인해야 합니다. Flask는 기본적으로 `templates`라는 디렉토리를 템플릿 파일이 위치할 기본 디렉토리로 인식합니다. 따라서 `login.html` 파일을 `templates` 디렉토리에 위치시켜야 합니다.

프로젝트 구조는 아래와 같아야 합니다:

```markdown

- 프로젝트 폴더
  - app.py (또는 다른 파일 이름)
  - templates
    - login.html
```



위의 구조로 파일을 배치한 후 다시 실행해 보세요. 그래도 `render_template('login.html')`에서 에러가 발생한다면, Flask 애플리케이션의 `templates` 디렉토리를 올바르게 설정했는지 확인해야 합니다. Flask 애플리케이션의 `__name__` 속성을 기준으로 `templates` 디렉토리를 설정하면 됩니다.

예를 들어, 아래와 같이 `app.py` 파일이 있다고 가정해 보겠습니다:

```python

from flask import Flask, render_template

app = Flask(__name__)

# ...

if __name__ == '__main__':
    app.run(debug=True)
```



이 경우, `__name__` 속성을 기준으로 `templates` 디렉토리를 설정하기 위해 아래와 같이 코드를 수정합니다:

```python

import os
from flask import Flask, render_template

app = Flask(__name__)

# templates 디렉토리 경로 설정
template_dir = os.path.abspath('templates')
app = Flask(__name__, template_folder=template_dir)

# ...

if __name__ == '__main__':
    app.run(debug=True)
```



위의 코드에서는 `template_dir` 변수를 사용하여 `templates` 디렉토리의 절대 경로를 얻고, `template_folder` 인자를 사용하여 `app` 객체를 생성할 때 `template_dir`을 전달합니다. 이렇게 하면 Flask 애플리케이션이 템플릿 파일을 올바르게 찾을 수 있게 됩니다.
	

1. 예제실행 

````python
from flask import Flask, render_template, request, redirect, url_for
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)   #starts ngrok when the app is run

# 로그인 페이지 렌더링
@app.route('/', methods=['GET', 'POST'])
def login():
    if request.method == 'POST':
        username = request.form['username']
        password = request.form['password']
        # 여기에 사용자 인증 로직 추가
        if username == 'admin' and password == 'password':
            return redirect(url_for('home'))
        else:
            error = 'Invalid username or password.'
            return render_template('login.html', error=error)
    return render_template('login.html')

# 홈 페이지
@app.route('/home')
def home():
    return "Welcome to the home page!"

if __name__ == '__main__':
    app.run()
````
