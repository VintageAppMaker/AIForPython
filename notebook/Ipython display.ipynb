1. `curses`: curses 라이브러리는 터미널 환경에서 CUI를 만들 수 있는 가장 일반적인 라이브러리입니다. 이 라이브러리는 터미널 창을 제어하고 키 입력 및 마우스 이벤트를 처리할 수 있습니다. curses는 터미널 기반의 텍스트 UI를 만들기에 적합하며, 주피터 노트북 환경에서도 사용할 수 있습니다.
2. `urwid`: urwid는 curses 라이브러리를 기반으로 한 파이썬용 CUI 라이브러리입니다. urwid는 다양한 위젯을 제공하며, 터미널에서 사용자 인터페이스를 만들기에 편리한 기능을 제공합니다. 주피터 노트북에서 urwid를 사용하면 터미널 창 내에서 CUI를 구현할 수 있습니다.
3. `prompt_toolkit`: prompt_toolkit은 명령줄 인터페이스를 개발하기 위한 라이브러리로서, 다양한 기능과 커스터마이징 옵션을 제공합니다. 주피터 노트북 환경에서도 prompt_toolkit을 사용하여 CUI를 구현할 수 있습니다. prompt_toolkit은 키 입력 및 터미널 창 제어와 같은 작업을 단순화하고, 커맨드 히스토리, 자동 완성 기능 등을 제공합니다.

위의 라이브러리 중 선택하여 원하는 CUI를 구현할 수 있습니다. 각 라이브러리에는 다양한 예제와 튜토리얼이 있으니 해당 라이브러리의 문서를 참조하시면 도움이 될 것입니다.

```python

from IPython.display import display

def handle_input(text):
    if text.strip().lower() == 'hello':
        display('Hello, World!')
    elif text.strip().lower() == 'quit':
        display('Goodbye!')
    else:
        display("I'm sorry, I didn't understand that.")

def interact():
    while True:
        text = input('Enter a command: ')
        handle_input(text)

interact()
```



위의 예제에서는 `handle_input()` 함수를 사용하여 사용자 입력을 처리하고, `display()` 함수를 사용하여 출력을 보여줍니다. `interact()` 함수에서는 무한 루프를 돌며 사용자 입력을 받고 처리합니다.

Colab 노트북에서는 사용자 입력을 받기 위해 `input()` 함수를 사용하고, Rich Output을 보여주기 위해 `display()` 함수를 사용합니다. 이를 통해 기본적인 대화형 CUI를 만들 수 있습니다.

위의 예제들은 `ipywidgets` 라이브러리를 사용하여 다양한 상호작용 요소를 생성하고 출력하는 방법을 보여줍니다. 위젯의 상태나 값 변화에 따라 출력을 업데이트하거나 이벤트를 처리하는 등 다양한 기능을 사용할 수 있습니다.


````python
from IPython.display import display
from PIL import Image
from IPython.display import display, HTML
from IPython.display import display, Markdown
from IPython.display import display, YouTubeVideo
import ipywidgets as widgets

import pandas as pd


# 이미지 출력 - Shell을 삭제하면 삭제됨
#image_path = "a.png"
#image = Image.open(image_path)
#display(image)

# html 출력
html_content = "<h1>Hello, World!</h1><br><img src = 'https://vintageappmaker.github.io/images/logo.png'>"
display(HTML(html_content))


# 표 출력
data = {'Name': ['John', 'Jane', 'Mike'],
        'Age': [25, 30, 35]}
df = pd.DataFrame(data)
display(df)


# markdown 출력
link = '[OpenAI](https://openai.com)'
display(Markdown(link))


# 동영상 출력
video_id = 'dQw4w9WgXcQ'  # YouTube video ID
video = YouTubeVideo(video_id)
display(video)


# 위젯 출력 활용
def on_button_clicked(b):
    print("Button clicked!")

button = widgets.Button(description="Click Me!")
button.on_click(on_button_clicked)
display(button)

# 위젯활용 II
text = widgets.Text(placeholder='Enter your name')
button = widgets.Button(description='Greet')
output = widgets.Output()

def greet_user(b):
    with output:
        output.clear_output()
        print(f'Hello, {text.value}!')

button.on_click(greet_user)

display(text, button, output)
````
