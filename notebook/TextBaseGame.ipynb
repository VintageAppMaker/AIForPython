### 실행소스

````python
# 텍스트 기반 게임 엔진 예제

# 게임 맵 정의
game_map = {
    '시작점': {
        '설명': '당신은 어두운 방 안에 있습니다.',
        '동': '다음방',
        '북': '비밀의방',
        '서': '배고픈용',
        '남': '출구'
    },
    '다음방': {
        '설명': '이 방은 밝습니다.',
        '서': '시작점'
    },
    '비밀의방': {
        '설명': '이 방은 비밀스럽습니다.',
        '남': '시작점'
    },
    '배고픈용': {
        '설명': '용이 배고파서 당신을 먹어버렸습니다. 게임 오버!',
        '게임오버': True
    },
    '출구': {
        '설명': '축하합니다! 당신은 미로를 탈출했습니다.',
        '게임오버': True
    }
}

# 게임 상태 초기화
current_room = '시작점'
game_over = False

# 게임 루프
while not game_over:
    # 현재 방의 정보 출력
    print(game_map[current_room]['설명'])
    
    # 플레이어 입력 받기
    command = input('어떻게 할까요? ')
    
    # 입력 처리
    if command == 'quit':
        game_over = True
    elif command == 'go':
        direction = input('어느 방향으로 이동하시겠습니까? ')
        if direction in game_map[current_room]:
            current_room = game_map[current_room][direction]
        else:
            print('잘못된 방향입니다.')
    elif command == 'look':
        print('주변을 살펴봅니다.')
    else:
        print('잘못된 명령입니다.')

# 게임 종료 메시지 출력
print('게임을 종료합니다.')
````
