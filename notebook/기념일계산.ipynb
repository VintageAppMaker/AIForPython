- 아래는 Python으로 기념일을 계산하는 간단한 예제입니다. 이 예제에서는 datetime 모듈을 사용하여 현재 날짜와 기념일 사이의 차이를 계산합니다. 이 예제에서는 사용자로부터 원하는 기념일과 현재 날짜를 입력받습니다.

```python

from datetime import datetime

def calculate_anniversary():
    # 사용자로부터 기념일과 현재 날짜를 입력받습니다.
    anniversary = input("기념일을 입력하세요 (YYYY-MM-DD 형식): ")
    current_date = datetime.now().date()

    try:
        # 입력된 기념일과 현재 날짜를 datetime 객체로 변환합니다.
        anniversary_date = datetime.strptime(anniversary, "%Y-%m-%d").date()

        # 현재 날짜와 기념일 사이의 차이를 계산합니다.
        difference = anniversary_date - current_date

        # 결과를 출력합니다.
        print("현재 날짜:", current_date)
        print("기념일:", anniversary_date)
        print("기념일까지 남은 시간:", difference.days, "일")

    except ValueError:
        print("잘못된 날짜 형식입니다.")

# 기념일 계산 함수를 호출합니다.
calculate_anniversary()
```



이 코드는 사용자로부터 "YYYY-MM-DD" 형식으로 기념일을 입력받고, 현재 날짜와 기념일 사이의 차이를 계산하여 출력합니다. 날짜 형식이 잘못된 경우에는 오류 메시지를 출력합니다. 이 코드를 실행하면 기념일까지 남은 날짜를 확인할 수 있습니다.


### 실행소스

````python
from datetime import datetime

def calculate_anniversary():
    # 사용자로부터 기념일과 현재 날짜를 입력받습니다.
    anniversary = input("기념일을 입력하세요 (YYYY-MM-DD 형식): ")
    current_date = datetime.now().date()

    try:
        # 입력된 기념일과 현재 날짜를 datetime 객체로 변환합니다.
        anniversary_date = datetime.strptime(anniversary, "%Y-%m-%d").date()

        # 현재 날짜와 기념일 사이의 차이를 계산합니다.
        difference = anniversary_date - current_date

        # 결과를 출력합니다.
        print("현재 날짜:", current_date)
        print("기념일:", anniversary_date)
        print("기념일까지 남은 시간:", difference.days, "일")

    except ValueError:
        print("잘못된 날짜 형식입니다.")

# 기념일 계산 함수를 호출합니다.
calculate_anniversary()
````
