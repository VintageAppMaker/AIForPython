# 파일찾기

> 다음과 같이 ChatGPT(🤖)의 프롬프트를 요청했습니다.
---
```python
python console 화면에서

데이터:
    검색할 단어 = ""
    찾은단어수  = 0

함수 메뉴보이기():
    다음과 같이 화면보이기
    "1. 파일찾기"
    "2. 파일 안에서 찾기"

    1을 누르면:
        입력을 받고 검색할 단어에 집어넣는다.
        지금 폴더에서 검색할 단어와 같은 파일을 찾는다.
        찾으면:
            "찾았습니다." 출력하며 찾은단어수를 +1
        아니면:
            "못찾았습니다" 출력    

    2을 누르면:
        "아직 구현하지 못했습니다" 출력


함수 결과보이기()
    "찾은 단어수는" + 찾은단어수
    

시작:
    메뉴보이기()
    결과보이기()
```
----
결과는 아래와 같습니다.

````python
# 데이터 초기화
검색할_단어 = ""
찾은단어수 = 0

# 함수 정의
def 메뉴보이기():
    print("1. 파일찾기")
    print("2. 파일 안에서 찾기")

    선택 = int(input("메뉴를 선택하세요: "))

    if 선택 == 1:
        검색할_단어 = input("검색할 단어를 입력하세요: ")

        # 현재 폴더에서 검색
        # 파일을 찾으면
        # 찾은단어수를 +1
        # "찾았습니다." 출력
        # 찾지 못하면
        # "못찾았습니다." 출력

        print("아직 구현하지 못했습니다.")

    elif 선택 == 2:
        print("아직 구현하지 못했습니다.")

def 결과보이기():
    print("찾은 단어수는 " + str(찾은단어수))

# 시작
메뉴보이기()
결과보이기()
````

```python
위의 소스를 윈도우에서 실행할 때, 해당경로의 파일을 찾아주는 내용으로 추가구현해주세요
```
----
결과는 다음과 같습니다.

````python
import os

# 데이터 초기화
검색할_단어 = ""
찾은단어수 = 0

# 함수 정의
def 메뉴보이기():

    print("1. 파일찾기")
    print("2. 파일 안에서 찾기")

    선택 = int(input("메뉴를 선택하세요: "))

    global 찾은단어수
    if 선택 == 1:
        검색할_단어 = input("검색할 단어를 입력하세요: ")

        # 현재 폴더에서 검색
        현재_폴더 = os.getcwd()
        파일_리스트 = os.listdir(현재_폴더)

        for 파일 in 파일_리스트:
            if 검색할_단어 in 파일:
                찾은단어수 += 1
                print("파일을 찾았습니다: " + 파일)

        if 찾은단어수 == 0:
            print("파일을 찾지 못했습니다.")

    elif 선택 == 2:
        print("아직 구현하지 못했습니다.")

def 결과보이기():
    global 찾은단어수  # Declare the variable as global
    print("찾은 단어수는 " + str(찾은단어수))

# 시작
메뉴보이기()
결과보이기()
````
