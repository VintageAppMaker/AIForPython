```python
from ipywidgets import Button
from IPython.display import display
from geopy.geocoders import Nominatim
import folium

def on_button_click(b):
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.geocode("37.7749, -122.4194", exactly_one=True)
    
    if location:
        lat = location.latitude
        lon = location.longitude
        
        # 지도 생성
        m = folium.Map(location=[lat, lon], zoom_start=12)
        
        # 마커 추가
        folium.Marker([lat, lon], popup='Your Location').add_to(m)
        
        # 지도 표시
        display(m)
    else:
        print("위치를 찾을 수 없습니다.")

button = Button(description="Get My GPS Address")
button.on_click(on_button_click)
display(button)
```



위 코드를 실행하면 "Get My GPS Address"라는 버튼이 나타나며, 이 버튼을 클릭하면 현재 GPS 위치를 가져와 지도에 표시됩니다. 코드의 `geolocator.geocode()` 메서드의 인자로 원하는 GPS 좌표를 전달하여 사용자의 실제 GPS 좌표를 얻을 수 있습니다.
